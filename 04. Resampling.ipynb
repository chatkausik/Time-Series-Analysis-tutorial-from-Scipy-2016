```python
import pandas as pd
import numpy as np
```

```python
rng = pd.date_range('1/1/2011', periods = 72, freq = 'H')
ts = pd.Series(list(range(len(rng))), index = rng)
ts.head(13)
```

```python
t=pd.date_range('01-01-2011', periods=72, freq='H')
kc=pd.Series(data=range(72), index=t)
kc.head()
```

```python
converted = ts.asfreq('45Min', method = 'bfill')
converted.head()
```

```python
# Let's try the more flexible .resample()
ts.resample('2H').mean().head()
```

```python
# What's particularly useful is that we can use reample to event out irregular time series
irreg_ts = ts[list(np.random.choice(a = list(range(len(ts))), size = 10, replace = False))]
```

```python
irreg_ts
```

```python
irreg_ts.asfreq('D')
```
